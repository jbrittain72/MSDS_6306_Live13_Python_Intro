```python
# Introduction to numpy
import numpy

my_numpy = numpy.array([45.4, 44.2, 36.8, 35.1, 39.0, 60.0, 47.4, 41.1, 45.8, 35.6])
print (my_numpy)

print (numpy.mean(my_numpy))
print (numpy.std (my_numpy))

print (my_numpy[my_numpy<45])

print (numpy.max(my_numpy))
print (numpy.min(my_numpy))
```

```python
# Introduction to pandas
import pandas as pd

# Need to use 'r' before the string to translate to RAW string
iris = pd.read_csv(r'D:\Users\Jim\SMU\MSDS 6306 DOING DATA SCIENCE\Files\iris.csv', header=0)

print (iris.head())
iris.drop('Id', axis=1, inplace=True)
print (iris.head())

print (iris.tail())
iris_sub = iris.drop(iris[iris.Species!='Iris-setosa'].index)
print (iris_sub.tail())

print (iris.describe())

grouped = iris.groupby(iris.Species)
print (grouped.describe())

# The following completes iut in one line
#print (iris.groupby(iris.Species).describe())
```
